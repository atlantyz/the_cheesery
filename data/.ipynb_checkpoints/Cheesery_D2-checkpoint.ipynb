```python
cheese_data_df['price'] = cheese_data_df['price'].astype(float, errors='raise')
```

```python
# Group wine price into bins.
price_counts = cheese_data_df.price.value_counts()
price_counts
```

```python
# Create price bins
price_bins = [0, 5, 10, 15, 20, 30,]
cheese_data_df.groupby(pd.cut(cheese_data_df["price"], price_bins)).count()

price_bin_names = ["less than $5","$5-10","$10-15","$15-20", "$20-30"]

# Make a new column in dataframe for price bins.
cheese_data_df["price_bins"] = pd.cut(cheese_data_df["price"], price_bins, labels=price_bin_names)

# Check dataframe
cheese_data_df.tail()
```

```python
# Read in our data
clean_cheese_df = pd.read_csv("Data/cheese_data_ml.csv")
```

```python
# read in again
clean_cheese_ml = pd.read_csv("Data/clean_cheese_df.csv")
```

```python
clean_cheese_ml.isnull()
```

```python
# create a variable for features to be encoded 
cheese_ft = ["cheese", "price_bins", "flavor"]
```

```python
# encode categorical data for splitting using onehotencoder
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(clean_cheese_df[cheese_ft]))

# Rename encoded columns
encode_df.columns = enc.get_feature_names(cheese_ft)
encode_df.head()
```

```python
# Merge the two DataFrames together and drop the flavor and price_bins column
clean_cheese_ml = clean_cheese_df.merge(encode_df,left_index=True,right_index=True)
```

```python
# save this csv also
clean_cheese_df.to_csv("Data/clean_cheese_df.csv", index=False)
```

```python
# drop columns not used for ml model
ml_cheese = clean_cheese_ml.drop(columns=["cheese", "company", "city", "milk", "country", "origin", "classification", "texture", "flavor", "vegetarian", "color"])
```

```python
ml_cheese
```

```python
from sklearn.model_selection import train_test_split
# target - cheese names
y = ml_cheese["price_bins"].values
# Features - price, flavor
X = ml_cheese.drop(["price_bins"],axis=1).values
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python

knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# fit model
knn_model = knn.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
y_pred = knn.predict(X_test)
```

```python

    confusion_matrix(y_test, y_pred),
```

```python
acc_score = accuracy_score(y_test, y_pred)
```

```python
# Displaying results
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, y_pred))
```
